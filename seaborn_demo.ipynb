```python
import pandas as pd
import seaborn as sns
import seaborn.objects as so
```

```python
supercars = pd.read_csv('https://learn.sharpsightlabs.com/datasets/pdm/supercars.csv')
```

```python
(so.Plot(data = supercars
         ,x = 'horsepower'
         ,y = 'top_speed'
         )
   .add(so.Dot())
 )
```

```python
stocks = pd.read_csv("https://www.sharpsightlabs.com/datasets/amzn_goog_2000-01-01_to_2020-12-05.csv")
stocks.date = pd.to_datetime(stocks.date)
amazon_stock = stocks.query("stock == 'amzn'")
```

```python
(so.Plot(data = amazon_stock
         ,x = 'date'
         ,y = 'close'
         )
   .add(so.Line())
 )
```

```python
bank = pd.read_csv('https://learn.sharpsightlabs.com/datasets/pdm/bank.csv')
(so.Plot(data = bank
         ,x = 'education'
         ,y = 'balance'
         )
   .add(so.Bar(), so.Agg())
 )
```
